```python
import urllib.request as req
import bs4 as bs

for page in range(1,60):
  url = f"https://tabelog.com/tw/tokyo/rstLst/{page}/?SrtT=rt"
  #"https://tabelog.com/tw/tokyo/rstLst/{}/?SrtT=rt".format(page)
  response = req.urlopen(url)
  html = bs.BeautifulSoup(response)

  # find_all: 形態是個類似list的
  r_list = html.find_all("div", {"class":"list-rst__body"})
  for r in r_list:
      en_name = r.find("a", {"class":"list-rst__rst-name-target"})
      en_name_href = en_name["href"]
      en_name_text = en_name.text
      area = r.find("div", {"class":"list-rst__area-genre cpy-area-genre"}).text.strip() #strip:去掉兩端空白,上python.string.method找
      title = r.find("p", {"class":"list-rst__pr-title cpy-pr-title"})
      if title:
        title = title.text.strip()
      else:
        title = "-"
      rating = r.find("span", {"class":"c-rating__val c-rating__val--strong list-rst__rating-val"})
      if rating:
        rating = rating.text
      else:
        rating = "-"
      prices = r.find_all("span", {"class":"c-rating-v3__val"})
      lunch_price = prices[0].text
      dinner_price = prices[1].text
      holiday = r.find("span", {"class":"list-rst__holiday-text"}).text

      print(en_name_text)
      print(en_name_href)
      print('地址:',area)
      print('簡述:',title)
      print('評分:',rating)
      print('午間價格:',lunch_price)
      print('晚間價格:',dinner_price)
      print('定休日:',holiday)
      print("-" * 30)
```
