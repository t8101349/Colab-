```python
# 請列出 離職 這個欄位的資料
data['離職(Attrition)']
```

```python
# 請查詢並列出 已經離職 的員工
is_Attrition = data['離職(Attrition)'] == 'Yes'
data[is_Attrition]
```

```python
# 請增加一個新的欄位 離職_數值
# 並分別用 0 跟 1 來表示這個員工是否離職
data['離職_數值'] = data['離職(Attrition)'].map({'Yes': 1, 'No': 0})
data
```

```python
# 請計算 加班 是否會影響的離職率
#data[['加班(OverTime)','離職(Attrition)']]
#data['離職(Attrition)'] = data['離職(Attrition)'].replace('Yes', 1)
#data['離職(Attrition)'] = data['離職(Attrition)'].replace('No', 0)
#data['加班(OverTime)'] = data['加班(OverTime)'].replace('Yes', 1)
#data['加班(OverTime)'] = data['加班(OverTime)'].replace('No', 0)
OverTime_group = data.groupby('加班(OverTime)')['離職_數值']
OverTime_stats = OverTime_group.agg(total='size', left='sum').reset_index()

OverTime_stats['離職率(Attrition Rate)'] = OverTime_stats['left'] / OverTime_stats['total']
OverTime_stats
```

```python
# 請計算 教育程度 是否會影響的離職率
# data[['教育程度(Education)','離職(Attrition)']]

Education_group = data.groupby('教育程度(Education)')['離職_數值']
Education_stats = Education_group.agg(total='size', left='sum').reset_index()

Education_stats['離職率(Attrition Rate)'] = Education_stats['left'] / Education_stats['total']
Education_stats
```

```python
# 請計算 各部門 的離職率
# 可使用 GroupBy
# Department = data.groupby(['部門(Department)','離職(Attrition)'])
# Department['離職(Attrition)'].size() / Department['部門(Department)'].size()
Department_group = data.groupby('部門(Department)')['離職_數值']
Department_stats = Department_group.agg(total='size', left='sum').reset_index()

Department_stats['離職率(Attrition Rate)'] = Department_stats['left'] / Department_stats['total']
Department_stats
```

```python
# 請找出哪個職位的離職率最高？  ans: Sales
Department_stats.sort_values(by='離職率(Attrition Rate)', ascending=False)
```

```python
# 請找出教育程度跟月收入之間是否有關係？ 平均有關
#月收入(MonthlyIncome) 教育程度(Education)
Education_Income_group = data.groupby('教育程度(Education)')['月收入(MonthlyIncome)'].mean().reset_index(name = 'mean_income')
Education_Income_group
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(data=Education_Income_group, x='教育程度(Education)', y='mean_income', color='skyblue')
plt.title('教育程度與平均月收入')
plt.xlabel('教育程度(Education)')
plt.ylabel('平均月收入(Mean Monthly Income)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(data=data, x='教育程度(Education)', y='月收入(MonthlyIncome)', color='skyblue')
plt.title('教育程度與月收入')
plt.xlabel('教育程度(Education)')
plt.ylabel('月收入(Monthly Income)')
plt.show()
```

```python
# 請計算各個部門的人力支出總額跟占比，並畫出派餅圖
# matplotlib.pyplot.pie
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html
import matplotlib.pyplot as plt
Department_salary_group = data.groupby('部門(Department)')['月收入(MonthlyIncome)']
Department_salary_stats = Department_salary_group.agg(Department_total='sum').reset_index()
print(Department_salary_stats)
#Department_salary_stats['total'] = Department_salary_group['Department_total'].sum().reset_index()


# Get the department names for labels
labels = [f"{row['部門(Department)']}: {row['Department_total']}"
          for _, row in Department_salary_stats.iterrows()]
plt.pie(x='Department_total', data=Department_salary_stats, labels=labels,
        pctdistance=0.6, labeldistance=1.1, startangle=0, radius=1)
plt.show()


total_salary = Department_salary_stats['Department_total'].sum()
labels2 = [f"{row['部門(Department)']}: {row['Department_total']} ({(row['Department_total'] / total_salary) * 100:.1f}%)"
          for _, row in Department_salary_stats.iterrows()]
plt.pie(x='Department_total', data=Department_salary_stats, labels=labels2,
        pctdistance=0.6, labeldistance=1.1, startangle=0, radius=1)
plt.show()
```

```python
# 請計算各部門的現職員工總數
# 並用長條圖顯示
# matplotlib.pyplot.bar
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
import matplotlib.pyplot as plt
Department_group = data.groupby('部門(Department)')['離職_數值']
Department_stats = Department_group.agg(total='size', left='sum').reset_index()
Department_stats['now'] = Department_stats['total'] - Department_stats['left']

plt.bar(x = '部門(Department)', height = 'now', width=0.5, data=Department_stats, label='now')
```
