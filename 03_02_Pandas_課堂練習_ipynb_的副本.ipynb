```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/drive/MyDrive/vgsales.csv 的副本")
data.head()
```

```python
# 取出 Publisher, Name, Genre 欄位的資料
data[['Publisher','Name','Genre']]
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和
data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum(axis = 1)
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和，並存入新的欄位"總和"
data['total_sales'] = data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum(axis = 1)
data
```

```python
#找出發行商為Nintendo的遊戲名稱
isNintendo = data['Publisher'] == 'Nintendo'
data[isNintendo]['Name']
```

```python
data[isNintendo]['Name'].nunique()
#data[isNintendo]['Name'].unique().size
```

```python
#列出所有Wii的遊戲名稱
on_wii = data['Platform'] == 'Wii'
data[on_wii]['Name'].unique()
```

```python
#計算所有GBA遊戲的銷售總合
on_gba = data['Platform'] == 'GBA'
data[on_gba]['Global_Sales'].sum()
```

```python
#計算任天堂運動類遊戲的銷售總量
c1 = data['Publisher'] == 'Nintendo'
c2 = data['Genre'] == 'Sports'
data[c1 & c2]['Global_Sales'].sum()
```

```python
#列出北美地區銷售量最高的前五名遊戲
sort_by_na = data.sort_values(by = 'NA_Sales',ascending = False)
sort_by_na['Name'].head(5)
```

```python
#列出北美與歐洲地區銷售量總和最高的前五名遊戲
na_eu = data['NA_Sales'] + data['EU_Sales']
data['na_eu_sale'] = na_eu
sort_by_na_eu = data.sort_values(by = 'na_eu_sale',ascending = False)
sort_by_na_eu[:5]['Name']
```

```python
#找出總共有多少種不同的平台
data['Platform'].nunique()
```

```python
#寫程式計算不同平台的總銷售量
platform = data['Platform'].unique()
for p in platform:
  this_p = data[data['Platform'] == p]
  this_p_sales = this_p['Global_Sales'].sum(axis = 0)
  print(p,this_p_sales)
```

```python
# 使用GroupBy來完成上述的工作
group_data = data.groupby('Platform')
group_data['Global_Sales'].sum()
```
