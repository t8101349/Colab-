```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
# 使用Seaborn可以很簡單的繪製
sns.relplot(
    data=iris,
    x="sepal_length", y="sepal_width",hue="species")
```

```python
# 相同的功能如果只用Matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# 取得不同物種的名稱
species = iris['species'].unique()

# 設定顏色對應物種
colors = sns.color_palette("Set1", len(species))

# 根據物種繪製不同顏色的點
for i, sp in enumerate(species):
    species_data = iris[iris['species'] == sp]
    plt.scatter(species_data['sepal_length'], species_data['sepal_width'],
                label=sp, color=colors[i], alpha=0.7)

# 設置標題和軸標籤
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# 顯示圖例
plt.legend(title='Species')

# 顯示圖形
plt.show()
```

sns分四大類:
1. relplot() :如此命名是因為它旨在視覺化許多不同的統計關係。雖然散佈圖通常很有效，但用一條線更好地表示一個變數代表時間測量的關係
2. displot() :散布圖
3. catplot() :分類圖。這些圖提供了不同程度的粒度。在最精細的層面上，您可能希望透過繪製「群」圖來查看每個觀察結果：一個散點圖，可調整點沿分類軸的位置，使它們不會重疊
4. pairplot():具有更廣泛的視角：它分別顯示所有成對關係和每個變數的聯合分佈和邊際分佈

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.scatter(tips['total_bill'], tips['tip'], marker='.', alpha=0.9)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
#練習題
#繪出散佈圖 x為bill_length_mm, y為bill_depth_mm
penguins = sns.load_dataset("penguins")
penguins.head()
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style='smoker')
```

```python
#單純使用Matplotlib來實現上圖
import matplotlib.pyplot as plt
import pandas as pd
tips_smoker = tips[tips['smoker'] == 'Yes']
tips_nonsmoker = tips[tips['smoker'] == 'No']

plt.scatter(tips_smoker['total_bill'], tips_smoker['tip'], marker='o',color='b', alpha=0.9)
plt.scatter(tips_nonsmoker['total_bill'], tips_nonsmoker['tip'], marker='x',color='y', alpha=0.9)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
#練習題
#使用下面資料，仿造上圖來繪圖
penguins = sns.load_dataset("penguins")
penguins.head()
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species", style='species')
sns.relplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="sex", style='species')
```
