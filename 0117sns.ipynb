```python
# Import seaborn
import seaborn as sns

# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")
tips.head()
```

```python
# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
# Seaborn
sns.relplot(
    data=iris,
    x="sepal_length", y="sepal_width",hue="species")
```

```python
# Matplotlib
import matplotlib.pyplot as plt
import pandas as pd

# 
species = iris['species'].unique()

# 
colors = sns.color_palette("Set1", len(species))

# 
for i, sp in enumerate(species):
    species_data = iris[iris['species'] == sp]
    plt.scatter(species_data['sepal_length'], species_data['sepal_width'],
                label=sp, color=colors[i], alpha=0.7)

# 
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# 
plt.legend(title='Species')

# 
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.relplot(data=tips, x="total_bill", y="tip")  # dataframe
```

```python
import matplotlib.pyplot as plt
import pandas as pd


plt.scatter(tips['total_bill'], tips['tip'], marker='.', alpha=0.9)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
#
# xbill_length_mm, ybill_depth_mm
penguins = sns.load_dataset("penguins")
penguins.head()
sns.relplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="smoker", style='smoker')
```

```python
#Matplotlib
import matplotlib.pyplot as plt
import pandas as pd
tips_smoker = tips[tips['smoker'] == 'Yes']
tips_nonsmoker = tips[tips['smoker'] == 'No']

plt.scatter(tips_smoker['total_bill'], tips_smoker['tip'], marker='o',color='b', alpha=0.9)
plt.scatter(tips_nonsmoker['total_bill'], tips_nonsmoker['tip'], marker='x',color='y', alpha=0.9)
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```
