第二題：請用 4x4 的子圖分別畫出四種不同屬性兩兩對應的分布圖

```python

from sklearn.datasets import load_iris
from matplotlib.colors import ListedColormap

iris = load_iris()
features = iris.data.T
colors = ['red', 'green', 'blue']
cmap = ListedColormap(colors)
plt.figure(figsize=(12,12))
for i in range(4):
  for j in range(4):
    plt.subplot(4,4,j*4+i+1)
    if i == j:
      x1 = features[i][target==0]
      x2 = features[i][target==1]
      x3 = features[i][target==2]
      plt.hist(x1, alpha=0.3, color='r')
      plt.hist(x2, alpha=0.3, color='g')
      plt.hist(x3, alpha=0.3, color='b')
    else:
      plt.scatter(features[i], features[j], c=iris.target, cmap=cmap, alpha = 0.6)

    if i == 0:
      plt.ylabel(iris.feature_names[j])
    if j == 3:
      plt.xlabel(iris.feature_names[i])

handles = [plt.plot([], [], color=cmap(i), marker='o', linestyle='')[0] for i in range(len(iris.target_names))]
labels = iris.target_names
plt.figlegend(handles, labels, loc='center right', bbox_to_anchor=(1.1, 0.5), title="Iris Species")
plt.tight_layout()
plt.show()
```
