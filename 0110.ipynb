```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
features = iris.data.T
target = iris.target

for i in range(4):
  feature0 = features[i][target == 0]  # Corrected: Indexing instead of function call
  feature1 = features[i][target == 1]  # Corrected: Indexing instead of function call
  feature2 = features[i][target == 2]  # Corrected: Indexing instead of function call

  plt.hist(feature0, alpha=0.3, label=iris.target_names[0])  # Corrected: Plotting feature0 data
  plt.hist(feature1, alpha=0.3, label=iris.target_names[1])  # Corrected: Plotting feature1 data
  plt.hist(feature2, alpha=0.3, label=iris.target_names[2])  # Corrected: Plotting feature2 data

  plt.xlabel(iris.feature_names[i])  # Adding x-axis label with feature name
  plt.ylabel("Frequency")  # Adding y-axis label
  plt.legend(loc = 4)  # Adding legend to identify species
  plt.show()
```
