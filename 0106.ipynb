```python
import urllib.request as req
import json

while True:
  url = 'https://www.cosmed.com.tw/api/getStore.aspx?t=store&c=&d=&s='
  f = req.urlopen(url)
  content = f.read()
  print(type(content))
  content = content.decode()
  if 'Invalid' in content:
    print('try...')
  else:
    print(content)
    print(type(content))
    break
```

```python
content = json.loads(content)
print(type(content))
file = open('cosmed.csv','w',encoding='utf-8')
file.write('店名,電話,地址\n')

store_list = content['data']
for store in store_list:
  print(store['StoreNM'])
  print(store['Tel'])
  print(store['Address'])
  print('-' * 10)
  file.write(store['StoreNM'])
  file.write(',')
  file.write(store['Tel'])
  file.write(',')
  file.write(store['Address'])
  file.write('\n')
file.close()
print('儲存')
```
